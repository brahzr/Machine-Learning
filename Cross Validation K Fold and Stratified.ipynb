Step 1: Importing necessary libraries

```python
from sklearn.model_selection import cross_val_score, KFold, StratifiedKFold
from sklearn.svm import SVC
from sklearn.datasets import load_iris
```

Step 2: Loading the dataset

```python
df = load_iris()
df
```

```python
iris = load_iris()
X, y = iris.data, iris.target
```

```python
y
```

Step 3: Creating SVM classifier

```python
svm_classifier = SVC(kernel='linear')
```

Step 4: Defining the number of folds for cross-validation

Step 5: Performing k-fold cross-validation & Stratified

```python
# ----- Normal K-Fold -----
kf = KFold(n_splits=5, shuffle=True, random_state=42)
results_kf = cross_val_score(svm_classifier, X, y, cv=kf)
```

```python
# ----- Stratified K-Fold -----
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
results_skf = cross_val_score(svm_classifier, X, y, cv=skf)
```

Step 6: Evaluation metrics

```python
# Print comparison
print("Normal K-Fold Results:", results_kf)
print("Mean Accuracy (KFold):", results_kf.mean())

print("\nStratified K-Fold Results:", results_skf)
print("Mean Accuracy (StratifiedKFold):", results_skf.mean())
```
