### Import libraries

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = fetch_california_housing()
type(df)
```

```python
dt = pd.DataFrame(df.data)
dt.columns = df.feature_names
dt.head()
```

```python
dt['MedHouseVal'] = df.target
```

```python
dt.head()
```

### Features and Target

```python
X = dt.iloc[:,:-1] #features
y = dt.iloc[:,-1] #target
```

```python
y.head()
```

### 1. Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import  KFold, StratifiedKFold
```

```python
linR = LinearRegression()
cv_score = cross_val_score(linR, X,y,cv = 5)
```

```python
print(cv_score)
```

```python
print("Cross Valodation Results:", cv_score)
print("Mean Accuracy (cv):", cv_score.mean())
```

```python
# ----- Normal K-Fold -----
skf = KFold(n_splits=5, shuffle=True, random_state=42)
results_kf = cross_val_score(linR, X, y, cv=skf)
```

```python
print(results_kf)
```

```python
print("Normal K-Fold Results:", results_kf)
print("Mean Accuracy (KFold):", results_kf.mean())
```

```python
#R_squared
from sklearn.metrics import r2_score
```

```python
linR.fit(X, y)
y_pred1 = linR.predict(X)
r2_score(y, y_pred1)
```

### Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

# GridSearchCV:
# Trains your model for each parameter combination.
# Uses cross-validation (splits data into folds).
# Evaluates performance (e.g., accuracy or MSE).

# It returns:
# The best parameters.
# The best score (e.g., accuracy).
# The best model fitted with those parameters.
```

```python
ridge = Ridge()
params = {
    "alpha": [1e-15, 1e-12, 1e-10, 1e-8, 1e-5, 5, 10, 20]
}
```

```python
ridg_reg = GridSearchCV(ridge,params, cv=5)
ridg_reg.fit(X,y)
```

```python
print(ridg_reg.best_params_)
print(ridg_reg.best_score_)
print(ridg_reg.best_estimator_)
```

```python
y_pred2 = ridg_reg.predict(X)
r2_score(y, y_pred2)
```

### Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
```

```python
lasso = Lasso()
params = {
    "alpha": [1e-15, 1e-12, 1e-10, 1e-8, 1e-5, 5, 10, 20]
}
```

```python
lass_reg = GridSearchCV(lasso,params, cv=5)
lass_reg.fit(X,y)
```

```python
print(lass_reg.best_params_)
print(lass_reg.best_score_)
print(lass_reg.best_estimator_)
print(lass_reg.feature_names_in_)
```

```python
y_pred3 = lass_reg.predict(X)
r2_score(y, y_pred3)
```
