## Gaussian Naive Bayes

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)
gnb = GaussianNB()
y_pred = gnb.fit(X_train, y_train).predict(X_test)
print("Number of mislabeled points out of a total", int(X_test.shape[0]), 
      "points:", int((y_test != y_pred).sum()))
```

## Multinomial Naive Bayes

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)
mnb = MultinomialNB()
y_pred = mnb.fit(X_train, y_train).predict(X_test)
print("Number of mislabeled points out of a total", int(X_test.shape[0]), 
      "points:", int((y_test != y_pred).sum()))
```

## Bernouli Naive Bayes

```python
import numpy as np
rng = np.random.RandomState(1)
X = rng.randint(5, size=(6, 100))
Y = np.array([1, 2, 3, 4, 4, 5])
from sklearn.naive_bayes import BernoulliNB #By default, BernoulliNB will binarize non-zero entries to 1,unless you change binarize=None.
clf = BernoulliNB(binarize=1.0)
#Any feature value greater than 1.0 is converted to 1
#Any feature value less than or equal to 1.0 is converted to 0
olf = BernoulliNB(binarize=0.0)
#Any feature value greater than 0.0 is converted to 1
#Any feature value equal to 0.0 remains 0
clf.fit(X, Y)
olf.fit(X,Y)
print(clf.predict(X[2:5]))
print(olf.predict(X[2:5]))
```
